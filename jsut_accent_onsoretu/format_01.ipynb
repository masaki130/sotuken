```python
import csv
import re

# 入力ファイル名と出力ファイル名
input_file = 'onsoretu.csv'
output_file = 'onsoretu_cleaned.csv'

# データの読み込みと変換
with open(input_file, 'r', encoding='utf-8') as infile, open(output_file, 'w', newline='', encoding='utf-8') as outfile:
    reader = csv.reader(infile)
    writer = csv.writer(outfile)

    for row in reader:
        if row:  # 空行をスキップ
            # 1つ目の要素から「BASIC5000_XXXX:」を削除し、必要な文字を削除
            cleaned_text = row[0].split(':', 1)[-1].replace('^', '').replace('#', '').replace('$', '').strip()
            # 半角の?を全角の？に変換
            cleaned_text = cleaned_text.replace('?', '？')
            # アンダースコアを削除
            cleaned_text = cleaned_text.replace('_', '')
            # 記号以外のカタカナをoに置き換え、伸ばし棒もoに置き換え
            cleaned_text = re.sub(r'[ァ-ン]', 'o', cleaned_text)  # カタカナをoに置き換え
            cleaned_text = cleaned_text.replace('ー', 'o')  # 伸ばし棒をoに置き換え

            # ？の直前の音節の直前に[を挿入し、？を削除
            cleaned_text = re.sub(r'(o)(？)', r'\[\1', cleaned_text)

            # クリーンなデータを書き込む
            writer.writerow([cleaned_text])

print(f'クリーンなデータが {output_file} に保存されました。')
```

```python
# onsoretu_cleaned.csv;カタカナ+韻律記号
# onsoretu_transformed.csv;カタカナ+数字
import os
import pandas as pd

# 対象のファイルパス
input_file_path = 'onsoretu_cleaned.csv'
output_file_path = 'onsoretu_transformed.csv'

def transform_onsoretu(input_file, output_file):
    # CSVファイルを読み込む
    df = pd.read_csv(input_file, header=None)
    
    transformed_lines = []
    
    # 各行を処理
    for index, row in df.iterrows():
        line = row[0]
        
        # [と]を削除し、対応する音素を変換
        transformed_line = ''
        i = 0
        while i < len(line):
            if line[i] == '[':
                # 次の音素を取得して1に変換
                i += 1
                if i < len(line):
                    transformed_line += '1' + line[i]  # 音素を1に変換
            elif line[i] == ']':
                # 次の音素を取得して2に変換
                i += 1
                if i < len(line):
                    transformed_line += '2' + line[i]  # 音素を2に変換
            elif line[i] not in '[]':
                # [ と ] 以外の文字をそのまま追加
                transformed_line += line[i]
            i += 1

        transformed_lines.append(transformed_line)
    
    # 変換した結果を新しいCSVファイルとして保存
    transformed_df = pd.DataFrame(transformed_lines)
    transformed_df.to_csv(output_file, index=False, header=False)

if __name__ == '__main__':
    transform_onsoretu(input_file_path, output_file_path)
    print(f"変換結果を '{output_file_path}' に保存しました。")
```
